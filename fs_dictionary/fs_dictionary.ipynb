```python
import pandas as pd
import nltk
```

```python
fs_dict = pd.read_csv('fpath to dictionary')
```

```python
fs_dict.head(20)
```

```python
fs_dict = fs_dict.fillna(-1).drop("Negative", axis=1).rename(columns={"Positive":"Sentiment"})
fs_dict.head()
```

```python
fs_dict.head(30)
```

```python
fs_dict.to_csv('path to transformed dictionary')
```

```python
fs_dict['Sentiment'].sum()
```

```python
sno = nltk.stem.SnowballStemmer('english')
fs_dict['Word'] = fs_dict['Word'].apply(sno.stem)

fs_dict = fs_dict.drop_duplicates()
```

```python
fs_dict.head(30)
```

```python
fs_dict.to_csv("path to stemmed dictionary")
```
